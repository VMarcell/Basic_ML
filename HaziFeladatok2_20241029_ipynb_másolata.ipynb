```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='target'), df['target'], test_size=0.2, random_state=42)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1);

factorial(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

clf_logreg = LogisticRegression(max_iter=10000, random_state=42)
clf_tree = DecisionTreeClassifier(random_state=42)
clf_knn = KNeighborsClassifier()

clf_logreg.fit(X_train, y_train)
clf_tree.fit(X_train, y_train)
clf_knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_logreg = clf_logreg.predict(X_test)
y_pred_tree = clf_tree.predict(X_test)
y_pred_knn = clf_knn.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_tree))
print("K-Nearest Neighbors Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
from sklearn.metrics import confusion_matrix

print("Logistic Regression Confusion Matrix:\n", confusion_matrix(y_test, y_pred_logreg))
print("Decision Tree Confusion Matrix:\n", confusion_matrix(y_test, y_pred_tree))
print("K-Nearest Neighbors Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

clf_logreg.fit(X_train_pca, y_train)
clf_tree.fit(X_train_pca, y_train)
clf_knn.fit(X_train_pca, y_train)

y_pred_logreg_pca = clf_logreg.predict(X_test_pca)
y_pred_tree_pca = clf_tree.predict(X_test_pca)
y_pred_knn_pca = clf_knn.predict(X_test_pca)

print("Logistic Regression (PCA) Accuracy:", accuracy_score(y_test, y_pred_logreg_pca))
print("Decision Tree (PCA) Accuracy:", accuracy_score(y_test, y_pred_tree_pca))
print("K-Nearest Neighbors (PCA) Accuracy:", accuracy_score(y_test, y_pred_knn_pca))
```
